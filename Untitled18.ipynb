# Анализ и обработка категориальных бинарных признаков

```python
# Проверим распределение количества по уникальным значениям

for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']:
    print (col)
    display(df[col].value_counts())
```

```python
# В бинарных параметрах 'nursery', 'higher', 'internet', 'schoolsup' пропущенные значения заполним значениями, которых более 80%.
# В столбцах 'famsup', 'paid', 'activities', 'romantic' пропущенные значения заполним рандомно, поскольку соотношение значений примерно 2/3

list_choice = ['yes', 'no']

for col in ['nursery', 'higher', 'internet']:
    df[col] = df[col].fillna('yes')
    
df.schoolsup = df.schoolsup.fillna('no')

for col in ['famsup', 'paid', 'activities', 'romantic']:
    df[col] = df[col].fillna(random.choice(list_choice))
```

```python
# school — аббревиатура школы, в которой учится ученик
# Категориальный номинативный признак, содержит два значения, пустых значений нет, выбросов нет.

display(df.school.value_counts())
df.school = df.school.apply(lambda x: 1 if x == 'GP' else 0) 
```

```python
# sex — пол ученика ('F' - женский, 'M' - мужской)
# Категориальный бинарный параметр, содержит два значения, пустых значений нет, выбросов нет.

display(df.sex.value_counts())
df.sex = df.sex.apply(lambda x: 1 if x == 'F' else 0) 
```

```python
# address — тип адреса ученика ('U' - городской, 'R' - за городом)
# Пустых значений 4%, адресс U - 75%.
# Предлагаем пустые значения заполнить значением 'U' и перевести значения в 1 и 0

display(df.address.value_counts())
df.address = df.address.fillna('U')
df.address = df.address.apply(lambda x: 1 if x == 'U' else 0) 
```

```python
# famsize — размер семьи('LE3' <= 3, 'GT3' >3)
# Пустых значений 7%.
# Предлагаем пустые значения заполнить значением 'GT3', как наиболее чаще встречающимся, и перевести значения в 1 и 0

display(df.famsize.value_counts())
df.famsize = df.famsize.fillna('GT3')
df.famsize = df.famsize.apply(lambda x: 1 if x == 'GT3' else 0) 
```

```python
# Pstatus — статус совместного жилья родителей ('T' - живут вместе 'A' - раздельно)
# Пустых значений 11,5%, раздельно живущих - 9, а совместно живущих - 79,5%. 
# Предлагаем заполнить пустые значения значением T, как наиболее часто встречающимся, и перевести значения в 1 и 0

display(df.Pstatus.value_counts())
df.Pstatus = df.Pstatus.fillna('T')
df.Pstatus = df.Pstatus.apply(lambda x: 1 if x == 'T' else 0) 
```

```python
# После всех проверок бинарных признаков, произведем замену yes на 1 и no на 0.

for col in ['nursery', 'higher', 'schoolsup', 'internet','famsup', 'paid', 'activities', 'romantic']:
    df[col] = df[col].apply(lambda x: 1 if x == 'yes' else 0)  
```

```python
# По боксплотам по влиянию бинарных переменных на целевую похоже, что параметры  'sex','school','famsize', 'Pstatus', 
#'paid', 'activities', 'famsup', 'nursery', имеют очень слабое влияние, а параметры 'schoolsup', 'address', 'higher', 
# 'internet','romantic' заметно влияют на оценку.

def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=df.loc[df.loc[:, column].isin(df.loc[:, column].value_counts().index.values)], 
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()

for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']:
    get_boxplot (col)
    
    
#    top_platforms = df.Platform.value_counts().sort_values(ascending = False).head(5).index.values
#sns.boxplot(y="Platform", x="Critic_Score", data=df[df.Platform.isin(top_platforms)], orient="h")
```

```python
# Проверим, есть ли статистическая разница в распределении оценок по категориальным признакам среди бинарных значений, 
# с помощью теста Стьюдента  - статистически значимые различия для колонок: sex,address, schoolsup, higher, romantic.

def get_student(column):
# тест Стьюдента

    cols = df.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
#    display (combinations_all)
    for comb in combinations_all:
        if ttest_ind(df.loc[df.loc[:, column] == comb[0], 'score'], 
                        df.loc[df.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']:
    get_student(col)
    
```

```python
# Проверим корреляцию между признаками с помощью коэффициента Мэтьюса
# Как видим, колонки друг с другом слабо коррелированы. Наибольшая корреляция есть между признаками school и address=0,26,
# schoolsup и school и sex = 0,14, internet и address = 0,2, higher и paid = 0,15 и т.д. 
# У параметров address, higher, romantic - корреляция друг с другом слабая.

from sklearn.metrics import matthews_corrcoef

list_columns = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']
matrix_metus = pd.DataFrame() #columns = ['school', 'sex'])

def get_metus(column):
# тест Мэтьюса

    for i in list_columns:
        matrix_metus.loc[i, column] = matthews_corrcoef(df[column], df[i])
       # matrix_metus[column][i] = ['2']

for col in list_columns:
    get_metus(col)
    
matrix_metus
```

```python
# guardian — опекун ('mother' - мать, 'father' - отец, 'other' - другое)
# Заполнили по-умолчанию наиболее частым значением, выбросов нет.

dic_guardian = {'mother': 0, 'father': 1, 'other': 2}

display(df.guardian.value_counts())
df.guardian=df.guardian.fillna('mother')
df.guardian = df.guardian.map(dic_guardian) 
```

```python
# Medu — образование матери (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее)
# Fedu — образование отца (0 - нет, 1 - 4 класса, 2 - 5-9 классы, 3 - среднее специальное или 11 классов, 4 - высшее)
# Смотрим, отличаются ли образование отца и матери в семье. 80% - это записи, где отец и мать имеют либо 
# идентичное образование, либо отличающееся на одну ступень. 
# Наилучшим вариантом заполнить пропуски видится - установить значение, равное образованию второго супруга, 
# или же создать отдельную колонку наибольшего значения образования по одному из родителей. Реализуем 1й вариант.

df.Fedu = df.Fedu.replace(40,4)  # Выброс исправляем

sum_same = 0
sum_1 = 0
for edu in range(0,len(df['Medu'])):
    if df['Medu'][edu] == df['Fedu'][edu]:
        sum_same +=1
    elif abs(df['Medu'][edu] - df['Fedu'][edu]) == 1:
        sum_1 += 1

print ("Родители с одинаковым образованием: ", sum_same, "\nРодители с образованием, отличным на одну ступень: ", sum_1, 
       "\nСреднее по столбцу: ", df['Medu'].mean(), df['Fedu'].mean(),  "\nМедиана: ", df['Medu'].median(), df['Fedu'].median())

for edu in range(0,len(df['Medu'])):
     if pd.isna(df['Medu'][edu]):
        df['Medu'][edu] = df['Fedu'][edu]    

for edu in range(0,len(df['Medu'])):
     if pd.isna(df['Fedu'][edu]):
        df['Fedu'][edu] = df['Medu'][edu]    
      
print ("После заполнения пропусков: \nСреднее по столбцу: ", df['Medu'].mean(), df['Fedu'].mean(),  "\nМедиана: ", df['Medu'].median(), df['Fedu'].median())
```

```python
# Mjob — работа матери, Fjob — работа отца  ('teacher' - учитель, 'health' - сфера здравоохранения, 
# 'services' - гос служба, 'at_home' - не работает, 'other' - другое)
# Пропущенные значения работы отца и матери, 36 и 19 соттветственно, заполним значением 'other', поскольку это наиболее часто встречающееся значение
# и наиболее подходящее в ситуации неопределенности

dic_job = {'teacher': 0, 'health': 1, 'services': 2, 'at_home': 3, 'other': 4}

display(df.Mjob.value_counts())
df.Fjob = df.Fjob.fillna('other')
df.Mjob = df.Mjob.fillna('other')
df.Fjob = df.Fjob.map(dic_job)
df.Mjob = df.Mjob.map(dic_job)
```

```python
# Матрица корреляций Пирсона.
# Согласно матрице коррекляций параметры слабо скоррелированы с целевой переменной. 
# Наиболее скоррелированные с целевой переменной: age.
# Между собой признаки слабо скорреллированные.

correlation = df_corr.corr()
cmap = sns.diverging_palette(5, 250, as_cmap=True)

def magnify():
    return [dict(selector="th",
                 props=[("font-size", "7pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "12pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '200px'),
                        ('font-size', '12pt')])
]

correlation.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Hover to magify")\
    .set_precision(2)\
    .set_table_styles(magnify())
```

```python
# По боксплотам заметно колебания значений целевой переменной для возраста 19-20 лет для age, в зависимости от absences 
# изменения целевой переменной заметны, оставляем колонку absences.

def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=df.loc[df.loc[:, column].isin(df.loc[:, column].value_counts().index[:])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()

for col in ['age', 'absences']:
    get_boxplot (col)
```
